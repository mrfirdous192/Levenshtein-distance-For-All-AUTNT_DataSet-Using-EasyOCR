```python
import cv2
import easyocr
import glob
import pytesseract
import pandas as pd
from nltk.metrics import edit_distance 
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('Test_Doc_Bengali.csv')
gt_string = df["RecognizedText"].tolist()
#gt_string

path = r"D:/AUDataset/MYDataSet/Bengali/Test_Doc_Bengali/img*.png"
i=0
Sum_Crr=0
dist_Test_Doc_Bengali=[0] *22
Sum_Crr=0
Distance1=[]

for file in sorted(glob.glob(path)):
    print(file)#Show Image File Address
    print('original=',gt_string[i])
    
    im= cv2.imread(file)
    reader = easyocr.Reader(['bn'])
    result = reader.readtext(im)
    if len(result)==0:
        i=i+1
        continue
    out_string = result[0][1]
    
    print('from image=',out_string)
    l_dist = edit_distance(gt_string[i],out_string)
    print('edit distance',l_dist)
    
    Distance1.append(l_dist)
    
    #Find Mean_crr:
    Total_character_length= len(gt_string[i])
    Dist_length=l_dist
    Crr=(Total_character_length-Dist_length)/Total_character_length
    Sum_Crr=Sum_Crr+Crr
    
    i=i+1
    dist_Test_Doc_Bengali[l_dist] = dist_Test_Doc_Bengali[l_dist]+1 
    print(dist_Test_Doc_Bengali[l_dist])
    
dist_Test_Doc_Bengali
```

```python
#find WRR
Wrr=(dist_Test_Doc_Bengali[0]/i)*100
print("Total Words =",i)
print("Distance in indexing =",dist_Test_Doc_Bengali)
print("Error free Words =",dist_Test_Doc_Bengali[0])
print("word recognition rate =",Wrr)

#find Mean_Crr
Mean_Crr=(Sum_Crr/i)*100
print("Mean Character Recognition rate =",Mean_Crr)
```

```python
import seaborn as sns
n_bins=10
sns.distplot(Distance1, n_bins,label='Latin', kde =True,rug=True,color="y",kde_kws = {'shade': False,'linewidth': 2})
```

```python
#plot histogram
plt.hist(Distance1, density=True,bins=10,range=(0,10), edgecolor="k", rwidth=0.8, color="#006747" )
plt.xlabel("Levenshtein distance")
plt.ylabel("Frequency")
plt.title("Distribution of MED for Bengali Test Document images")

plt.savefig('out_hist/Test_Doc_Bengali.png', bbox_inches='tight', quality=100)
plt.show()
```

```python
#BarGraph:
i=np.arange(22)
tick_label=i
plt.bar(i,dist_Test_Doc_Bengali,label ='Test_Doc_Bengali',tick_label = tick_label,width=0.8, color="#006747")

plt.xlabel("Levenshtein Distance")
plt.ylabel("No_Of_Image")
plt.title("Bar Chart of MED for Bengali Test Document images")

plt.savefig('out_bar/Test_Doc_Bengali.png', bbox_inches='tight', quality=100)
plt.show()
```
