```python
import cv2
import easyocr
import glob
import pytesseract
import pandas as pd
from nltk.metrics import edit_distance 
import numpy as np
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('GT_Test_Doc_Latin.csv')
gt_string = df["RecognizedText"].tolist()
gt_string

#define the variables
path = r"D:/AUDataset/MYDataSet/Latin/Test_Doc_Latin/img*.png"
i=0
dist_Test_Doc_Latin=[0] *22
Sum_Crr=0
Distance1=[]

for file in sorted(glob.glob(path)):
    #print(file)#Show Image File Address
    print('original=',gt_string[i])
    im= cv2.imread(file)
    reader = easyocr.Reader(['en'])
    result = reader.readtext(im)

    out_string = result[0][1]
    
    print('from image=',out_string)
    l_dist = edit_distance(gt_string[i],out_string)
    print('edit distance',l_dist)
    
    Distance1.append(l_dist)
    
    #Find Mean_crr:
    Total_character_length= len(gt_string[i])
    Dist_length=l_dist
    Crr=(Total_character_length-Dist_length)/Total_character_length
    Sum_Crr=Sum_Crr+Crr
    
    i=i+1
    dist_Test_Doc_Latin[l_dist] = dist_Test_Doc_Latin[l_dist]+1 
    print(dist_Test_Doc_Latin[l_dist])
    
dist_Test_Doc_Latin
```

```python
#find WRR
Wrr=(dist_Test_Doc_Latin[0]/i)*100
print("Total Words =",i)
print("Distance in indexing =",dist_Test_Doc_Latin)
print("Error free Words =",dist_Test_Doc_Latin[0])
print("word recognition rate =",Wrr)

#find Mean_Crr
Mean_Crr=(Sum_Crr/i)*100
print("Mean Character Recognition rate =",Mean_Crr)
```

```python
#plot histogram
plt.hist(Distance1, density=True,bins=10,range=(0,10), edgecolor="k", rwidth=0.8 )
plt.xlabel("Levenshtein distance")
plt.ylabel("Frequency")
plt.title("Distribution of MED for Latin Test document images")

plt.savefig('out_hist/Test_Doc_Latin.png', bbox_inches='tight', quality=100)
plt.show()
```

```python
#BarGraph:
i=np.arange(22)
tick_label=i
plt.bar(i,dist_Test_Doc_Latin,label ='Test_Scene_Latin',tick_label = tick_label,width=0.8)

plt.xlabel("Levenshtein_Distance")
plt.ylabel("No_Of_Image")
plt.title("Bar Chart of MED for Latin test document images")

plt.savefig('out_bar/Test_Doc_Latin.png', bbox_inches='tight', quality=100)
plt.show()
```
